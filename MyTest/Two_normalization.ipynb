```python
# 数据归一化
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
original_data = np.random.exponential(size=1000)
scaled_data = (original_data-original_data.min())/(original_data.max()-original_data.min())
# plt.subplots() 返回一个 Figure实例fig 和一个 AxesSubplot实例ax 。这个很好理解，fig代表整个图像，ax代表坐标轴和画的图。 
fig,ax = plt.subplots(1,2)
sns.distplot(original_data,ax=ax[0])
ax[0].set_title("original_data")
sns.distplot(scaled_data,ax=ax[1])
ax[1].set_title("scaled_data")
plt.show() 
```

归一化：对数据的数值范围进行特定缩放，但不改变其数据分布的一种线性特征变换。

```python
lamp = 6
original_data =np.random.poisson(lamp,size=1000)
normalized_data = (original_data-original_data.mean())/original_data.std()
fig,ax = plt.subplots(1,2)
sns.distplot(original_data,ax=ax[0])
ax[0].set_title("original_data")
sns.distplot(normalized_data,ax=ax[1])
ax[1].set_title("normalized_data")
```

标准化：对数据的分布的进行转换，使其符合某种分布（比如正态分布）的一种非线性特征变换。

```python
import scipy as cp
original_data = np.random.exponential(size=1000)
standardized_data = cp.stats.boxcox(original_data)[0]
fig,ax = plt.subplots(1,2)
sns.distplot(original_data,ax=ax[0])
ax[0].set_title("original_data")
sns.distplot(standardized_data,ax=ax[1])
ax[1].set_title("standardized_data")
plt.show()
```
