```python
import numpy as np
def iris_type(s):
    it={b'Iris-setosa':0,b'Iris-versicolor':1,b'Iris-virginica':2}
    return it[s]
path = '/Users/tequila2411/Desktop/Iris.data'
data = np.loadtxt(path,dtype=float,delimiter=',',converters={4:iris_type})
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
x,y = np.split(data,(4,),axis=1)
x = x[:,:2]
x_train,x_test,y_train,y_test =train_test_split(x,y,random_state=1,test_size=0.6)
print("x_train:",x_train)
print("x_test:",x_test)
print("y_train:",y_train)
print("y_test:",y_test)
```

```python
from sklearn import svm
clf = svm.SVC(C =0.8,kernel="rbf",gamma=20, decision_function_shape='ovr')
clf.fit(x_train,y_train.)
clf.score(x_train,y_train)
y_hat = clf.predict(x_train)
clf.score(x_test,y_test)
y_hat = clf.predict(x_test)
```

```python
x1_min, x1_max = x[:, 0].min(), x[:, 0].max()  # 第0列的范围
x2_min, x2_max = x[:, 1].min(), x[:, 1].max()  # 第1列的范围
x1, x2 = np.mgrid[x1_min:x1_max:200j, x2_min:x2_max:200j]  # 生成网格采样点
grid_test = np.stack((x1.flat, x2.flat), axis=1)  # 测试点
# print 'grid_test = \n', grid_testgrid_hat = clf.predict(grid_test)       # 预测分类值grid_hat = grid_hat.reshape(x1.shape)  # 使之与输入的形状相同
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import colors
 
cm_light = mpl.colors.ListedColormap(['#A0FFA0', '#FFA0A0', '#A0A0FF'])
cm_dark = mpl.colors.ListedColormap(['g', 'r', 'b'])
plt.pcolormesh(x1, x2, grid_hat, cmap=cm_light)
# plt.scatter(x1[:,0],x1[:,1], c='r', edgecolors='k', s=50, cmap=cm_dark)
plt.plot(x[:, 0], x[:, 1], 'o', alpha=0.5, color='blue', markeredgecolor='k')
plt.scatter(x_test[:, 0], x_test[:, 1], s=120, facecolors='none', zorder=10)  # 圈中测试集样本
plt.xlabel(u'longth', fontsize=13)
plt.ylabel(u'width', fontsize=13)
plt.xlim(x1_min, x1_max)
plt.ylim(x2_min, x2_max)
plt.title(u'SVM classifier', fontsize=15)
plt.show()
```
